```python
import matplotlib.pyplot as plotlib
import numpy as npy
```

inter-arrival time

```python
def int_arvl(n,arvl_rate):
    inter_arvl_time=[]
    for i in range(n):
        temp = int(npy.random.exponential(1/arvl_rate)*60*60)
        if i==0:
            inter_arvl_time.append(0)
        else:
            inter_arvl_time.append(temp)
    return inter_arvl_time
```

arrival time

```python
def arvl_time(n,inter_arvl_time):
    arvl_time=[]
    for i in range(n):
        if i==0:
            arvl_time.append(0)
        else:
            arvl_time.append(arvl_time[i-1]+inter_arvl_time[i])
    return arvl_time
```

serving Time

```python
def srv_time(n,srv_rate):
    srv_time=[]
    while not len(srv_time) == n:
        t = int(npy.random.exponential(1/srv_rate)*60*60)
        if not t<1:
            srv_time.append(t)
    return srv_time
```

```python
total_time=int(input("Please enter total amount of time in hours:"))
arrival_rate=int(input("Please enter arrival amount in hours:"))
service_rate=int(input("Please enter Service rate in hours: "))
```

```python
n=int((npy.random.poisson(arrival_rate))*total_time)
```
