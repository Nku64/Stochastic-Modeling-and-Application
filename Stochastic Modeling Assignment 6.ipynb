```python
import matplotlib.pyplot as plotlib
import numpy as npy
```

```python
#inter-arrival time
def int_arvl(n,arvl_rate):
    inter_arvl_time=[]
    for i in range(n):
        temp = int(npy.random.exponential(1/arvl_rate)*60*60)
        if i==0:
            inter_arvl_time.append(0)
        else:
            inter_arvl_time.append(temp)
    return inter_arvl_time
```

```python
#arrival time
def arvl_time(n,inter_arvl_time):
    arvl_time=[]
    for i in range(n):
        if i==0:
            arvl_time.append(0)
        else:
            arvl_time.append(arvl_time[i-1]+inter_arvl_time[i])
    return arvl_time
```

```python
#serving Time
def srv_time(n,srv_rate):
    srv_time=[]
    while not len(srv_time) == n:
        t = int(npy.random.exponential(1/srv_rate)*60*60)
        if not t<1:
            srv_time.append(t)
    return srv_time
```

```python
def statistics(n,srv_rate,arvl_rate):
    Lembda=arvl_rate/srv_rate
    pi=[]
    for j in range (n):
        p=(1-Lembda)*(Lembda**j)
        pi.append(p)
    
    L_s=Lembda/(1-Lembda)
    L_q=L_s-Lembda
    W_s=L_s/arvl_rate
    W_q = W_s - (1/srv_rate)
    
    return pi,L_s,L_q,W_s,W_q
```

```python
n=int((npy.random.poisson(arrival_rate))*total_time)
```

```python
inter_arrival_time=int_arvl(n,arrival_rate)
arrival_time=arvl_time(n,inter_arrival_time)
service_time=srv_time(n,service_rate)
```

```python
#taking values 
pi,L_s,L_q,W_s,W_q=statistics(n,service_rate,arrival_rate)
```
